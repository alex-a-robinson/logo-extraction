# Method 1 - Grab cut, with drawn bounding box
Use the grabcut algorithm with a drawn bounding box around the image.

## Anaylsis
 + Removes some of background
 - Dosen't remove all of background
 - Requires drawn bounding box

```python
img = cv2.imread('data/firefox_logo_non_croped.jpg')

original = img.copy()
cv2.rectangle(original, (150, 35), (250+150, 245+35), (255,0,0), 2)


plt.figure()
plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))

mask = np.zeros(img.shape[:2],np.uint8)

bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)

rect = (150,35,250, 245)

cv2.grabCut(img,mask,rect,bgdModel,fgdModel,5,cv2.GC_INIT_WITH_RECT)

mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
img = img*mask2[:,:,np.newaxis]


plt.figure()
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

# Method 2 - Feature detection and minimum bounding box
Detect the logo's feature and draw a minimum bounding box, then remove everything outside of the box

```python
img = cv2.imread('data/firefox_logo_non_croped.jpg')

original = img.copy()

plt.figure()
plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))

# Get grayscale image
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply slight blur to reduce noise
blur = cv2.bilateralFilter(gray, 11, 17, 17)

# Apply threshold on white background
ret, thresh = cv2.threshold(blur, 250, 255, cv2.THRESH_BINARY_INV)

plt.figure()
plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_GRAY2RGB))

# Mask orginal image from threshold
masked = cv2.bitwise_and(img, img, mask = thresh)

plt.figure()
plt.imshow(cv2.cvtColor(masked, cv2.COLOR_BGR2RGB))

# Find the contours from our threshold mask and pick the one with the largest area
contours, _ = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
largest_contour = sorted(contours, key = cv2.contourArea, reverse = True)[0]

# Find its minimum bounding box
x, y, w, h = cv2.boundingRect(largest_contour)

# Crop the image
roi = masked[y:y+h,x:x+w]

plt.figure()
plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))

plt.show()
```

```python
# Wrap the above methods into a nice function

def logo_cleanup(img):
    # Get grayscale image
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Apply slighy blur to reduce noise
    blur = cv2.bilateralFilter(gray, 11, 17, 17)

    # Apply threshold on white background
    ret, thresh = cv2.threshold(blur, 250, 255, cv2.THRESH_BINARY_INV)

    # Mask orginal image from threshold
    masked = cv2.bitwise_and(img, img, mask = thresh)
    
    # Find the contours from our threshold mask and pick the one with the largest area
    contours, _ = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
    largest_contour = sorted(contours, key = cv2.contourArea, reverse = True)[0]

    # Find its minimum bounding box
    x, y, w, h = cv2.boundingRect(largest_contour)

    # Crop the image
    roi = masked[y:y+h,x:x+w]

    return roi
```

```python
def logo_extract(path):
    img = cv2.imread(path)

    clean_img = logo_cleanup(img.copy())

    plt.figure()
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.figure()
    plt.imshow(cv2.cvtColor(clean_img, cv2.COLOR_BGR2RGB))

    plt.show()
```

```python
import urllib
import os

def download(url, name, extension):
    '''Downloads and saves an image if it dosen't already exsist
    '''
    if not os.path.isfile('data/%s.%s' % (name, extension)):
        with open('data/%s.%s' % (name, extension), 'wb') as image_file:
            image_file.write(urllib.urlopen(url).read())
            image_file.close()
```
